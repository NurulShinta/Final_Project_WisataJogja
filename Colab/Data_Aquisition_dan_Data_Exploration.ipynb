### scraping web yogyes.com

Scraping web dengan menggunakan beautifulsoup untuk website https://www.yogyes.com/id/yogyakarta-tourism-object/nature-and-outdoor karena ada informasi mengenai wisata alam di jogja tetapi tidak tahu nantikan akan digunakan atau tidak data ini

```python
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
from csv import writer

url = "https://www.yogyes.com/id/yogyakarta-tourism-object/nature-and-outdoor/"
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
lists = soup.find_all('div', class_='table-cell')

with open('wisata_web.csv', 'w', encoding='utf8', newline='') as f:
    thewriter =  writer(f)
    header = ['Nama Wisata', 'Deskripsi Wisata']
    thewriter.writerow(header)


    for list in lists:
      name = list.find('span', class_="name").text.replace('\n', '')
      description = list.find('p', class_="description").text.replace('\n', '')
      info = [name, description]
      thewriter.writerow(info)
      print(info)
```

```python
# menghapus kolom yang tidak diperlukan dan menggabungkan
df1 = pd.read_csv('/content/Datawisata1.csv')
df1 = df1.drop(['Maps', 'Unnamed: 0', 'Unnamed: 0.1', 'Maps', 'harga_tiket', 'Alamat', 'Review', 'Deskripsi Wisata', 'Min Harga Tiket', 'Max Harga Tiket'],axis = 1)
df2 = pd.read_csv('/content/Datawisata2.csv')
df2 = df2.drop(['no', 'sumber_tiket', 'htm_weekday', 'htm_weekend', 'latitude', 'longitude', 'Deskripsi Wisata', 'harga_tiket'],axis = 1)
frames = [df1, df2]
df_all = pd.concat(frames).drop_duplicates(subset='Nama Wisata', keep='first').reset_index(drop=True)
```

```python
# cek missing value
df_all.isnull().sum()
```

```python
# mengubah type data
df = df_all
df["Rating"] = df["Rating"].str.replace(',', '.').astype(float)
df["Ulasan"] = df["Ulasan"].astype(str).str.replace('.,', '').astype(int)
df["rata2_tiket"] = df["rata2_tiket"].astype(float).astype(int)
df
```

```python

temporary = df['type']
df.drop(labels=['type'], axis=1, inplace = True)
df['Type_Wisata'] = temporary # pindahkan label ke posisi ujung sebelah kanan
df.head()
```

```python
# cek data unik type wisata
print('type ', df['Type_Wisata'].nunique())
print( '\n', df['Type_Wisata'].unique())
```
